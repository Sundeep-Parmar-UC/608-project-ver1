```python
import requests
import zstandard as zstd
import os

def download_and_decompress_zst(url, zst_filepath, output_filepath):
    """
    Downloads a ZST file and decompresses it.

    :param url: The URL of the .zst file to download.
    :param zst_filepath: Local path to save the downloaded ZST file.
    :param output_filepath: Local path for the decompressed output file.
    """
    # 1. Download the file
    print(f"Downloading {url} to {zst_filepath}...")
    try:
        response = requests.get(url, stream=True)
        response.raise_for_status() # Ensure the download was successful

        with open(zst_filepath, 'wb') as f:
            for chunk in response.iter_content(chunk_size=None):
                f.write(chunk)
        print("Download complete.")

    except requests.exceptions.RequestException as e:
        print(f"Download failed: {e}")
        return

    # 2. Decompress the file
    print(f"Decompressing {zst_filepath} to {output_filepath}...")
    try:
        dctx = zstd.ZstdDecompressor()
        with open(zst_filepath, 'rb') as ifh, open(output_filepath, 'wb') as ofh:
            # Use copy_stream for efficient, memory-friendly streaming decompression
            dctx.copy_stream(ifh, ofh)
        print("Decompression complete.")

    except Exception as e:
        print(f"Decompression failed: {e}")

    # Optional: Remove the compressed file after decompression
    # os.remove(zst_filepath)
    # print(f"Removed compressed file {zst_filepath}")

# Example Usage:
# Replace with an actual URL to a .zst file
example_url = "https://database.lichess.org/standard/"+ NeededFile
example_zst_path = NeededFile
example_output_path = NeededFile[:-4]
download_and_decompress_zst(example_url, example_zst_path, example_output_path)

#https://database.lichess.org/standard/lichess_db_standard_rated_2013-02.pgn.zst
#https://database.lichess.org/standard/lichess_db_standard_rated_2013-02.pgn.zst
#flask_app/venv/bin/fourth.py
```

```python
#set timer start
TimeStart = time.time()

#load in the chess data

#------------------------------
#define url to github and assess to smaller SampleData
#file_url = "https://raw.githubusercontent.com/Sundeep-Parmar-UC/Project1/refs/heads/main/sampleData.txt"
#file_url = "flask_app/venv/bin/" + example_output_path
#lichess_db_standard_rated_2013-02.pgn
#define url to github and assess to main Project Data set
#file_url = "https://raw.githubusercontent.com/Sundeep-Parmar-UC/Data606-Project/refs/heads/main/Citizen_Satisfaction_Survey_(2016-2017)_20260126.csv"
#------------------------------

#Set variable to stop collecting events after # lines read
Stop_loading_data_after_lines = 10000

# Define the desired column names
EventArray = ['Event','Site','White','Black','Result','UTCDate','UTCTime','WhiteElo','BlackElo','WhiteRatingDiff','BlackRatingDiff','ECO','Opening','TimeControl','Termination','Moves','First_Move','Second_Move','Third_Move','White_Move_Count','Black_Move_Count']
## create chess games data frame
chessg = pd.DataFrame(columns=EventArray)

LinesTotalRead = 0 #line number of input data
EventInfoGathered = 0 #number of event elements collected
MoveLineGathered = False # indicator of last event element collected
Eventscollected = 0 #number of events collected

#define filename of local data set
local_filename = example_output_path

try:
    # Download the file from the URL and save it locally
    #urllib.request.urlretrieve(file_url, local_filename)
    #print(f"File '{local_filename}' downloaded successfully from '{file_url}'.")

    #open local copy of the data set
    with open(local_filename, 'r') as file:
        while True:
            if LinesTotalRead == Stop_loading_data_after_lines:
              print("Data collection stopped at line: "+str(Stop_loading_data_after_lines))#leave message and stop loop
              break

            line = file.readline() #read one line from data file
            LinesTotalRead+=1 # increment lines read
            if not line:  # Check if the line is an empty string, indicating EOF
                print("End of file reached.") # stop processing, all events should have been saved.
                break
            elif line.startswith('[Event '): #Found Event info
                #print("Event Info found.")
                EventInfoGathered+=1
                EventData = line[8:(9+line[9:].find("\""))]
                #print(EventData)
            elif line.startswith('[Site '): #Found Site info
                #print("Site Info found.")
                EventInfoGathered+=1
                SiteData = line[7:(8+line[8:].find("\""))]
                #print(SiteData)
            elif line.startswith('[White '): #Found White info
                #print("White Info found.")
                EventInfoGathered+=1
                WhiteData = line[8:(9+line[9:].find("\""))]
                #print(WhiteData)
            elif line.startswith('[Black '): #Found Black info
                #print("Black Info found.")
                EventInfoGathered+=1
                BlackData = line[8:(9+line[9:].find("\""))]
                #print(BlackData)
            elif line.startswith('[Result '): #Found Result info
                #print("Result Info found.")
                EventInfoGathered+=1
                ResultData = line[9:(10+line[10:].find("\""))]
                #print(ResultData)
            elif line.startswith('[UTCDate '): #Found UTCDate info
                #print("UTCDate Info found.")
                EventInfoGathered+=1
                UTCDateData = line[10:(11+line[11:].find("\""))]
                #print(UTCDateData)
            elif line.startswith('[UTCTime '): #Found UTCTime info
                #print("UTCTime Info found.")
                EventInfoGathered+=1
                UTCTimeData = line[10:(11+line[11:].find("\""))]
                #print(UTCTimeData)
            elif line.startswith('[WhiteElo '): #Found WhiteElo info
                #print("WhiteElo Info found.")
                EventInfoGathered+=1
                WhiteEloData = line[11:(12+line[12:].find("\""))]
                #print(WhiteEloData)
            elif line.startswith('[BlackElo '): #Found BlackElo info
                #print("BlackElo Info found.")
                EventInfoGathered+=1
                BlackEloData = line[11:(12+line[12:].find("\""))]
                #print(BlackEloData)
            elif line.startswith('[WhiteRatingDiff '): #Found WhiteRatingDiff info
                #print("WhiteRatingDiff Info found.")
                EventInfoGathered+=1
                WhiteRatingDiffData = line[18:(19+line[19:].find("\""))]
                #print(WhiteRatingDiffData)
            elif line.startswith('[BlackRatingDiff '): #Found BlackRatingDiff info
                #print("BlackRatingDiff Info found.")
                EventInfoGathered+=1
                BlackRatingDiffData = line[18:(19+line[19:].find("\""))]
                #print(BlackRatingDiffData)
            elif line.startswith('[ECO '): #Found ECO info
                #print("ECO Info found.")
                EventInfoGathered+=1
                ECOData = line[6:(7+line[7:].find("\""))]
                #print(ECOData)
            elif line.startswith('[Opening '): #Found Opening info
                #print("Opening Info found.")
                EventInfoGathered+=1
                OpeningData = line[10:(11+line[11:].find("\""))]
                #print(OpeningData)
            elif line.startswith('[TimeControl '): #Found TimeControl info
                #print("TimeControl Info found.")
                EventInfoGathered+=1
                TimeControlData = line[14:(15+line[15:].find("\""))]
                #print(TimeControlData)
            elif line.startswith('[Termination '): #Found Termination info
                #print("Termination Info found.")
                EventInfoGathered+=1
                TerminationData = line[14:(15+line[15:].find("\""))]
                #print(TerminationData)
            elif line.startswith('1. '): #Found Moves info
                #print("Moves Info found.")
                EventInfoGathered+=1
                MovesData = line.strip() # Use the entire line for the Moves column

                MovesData = re.sub(r"\{ \[%eval \d\.\d\d\] \}", "", MovesData)
                MovesData = re.sub(r"\{ \[%clk \d\.\d\d\] \}", "", MovesData)
                MovesData = MovesData.replace("...", ".")


                # Use regex to extract moves and move counts
                moves = re.findall(r'\d+\.\s+([^\s]+)\s+([^\s]+)', MovesData)
                # Extract first three moves
                FirstMoveData = moves[0][0] if len(moves) > 0 else ''
                SecondMoveData = moves[0][1] if len(moves) > 0 else ''
                if SecondMoveData == "1.":
                  SecondMoveData = MovesData[MovesData.find("1.",5)+2:MovesData.find("1.",5)+5]

                ThirdMoveData = moves[1][0] if len(moves) > 1 else ''

                # Extract move counts
                move_counts = re.findall(r'(\d+)\.\s+[^\s]+', MovesData)
                WhiteMoveCount = move_counts[-1] if len(move_counts) > 0 else ''
                BlackMoveCount = str(int(WhiteMoveCount) - 1) if WhiteMoveCount and MovesData.endswith('...') else WhiteMoveCount


                MoveLineGathered = True
                EventInfoGathered+=1

            #All data for single event has been obtained. Store in main data frame
            if (MoveLineGathered):
                MoveLineGathered = False
                #print("Single Event Data captured")

                #EventArray = ['Event','Site','White','Black','Result','UTCDate','UTCTime','WhiteElo','BlackElo','WhiteRatingDiff','BlackRatingDiff','ECO','Opening','TimeControl','Termination','Moves','First_Move','Second_Move','Third_Move','White_Move_Count','Black_Move_Count']
                chessg.loc[Eventscollected] = [EventData,SiteData,WhiteData,BlackData,ResultData,UTCDateData,UTCTimeData,WhiteEloData,BlackEloData,WhiteRatingDiffData,BlackRatingDiffData,ECOData,OpeningData,TimeControlData,TerminationData,MovesData,FirstMoveData,SecondMoveData,ThirdMoveData,WhiteMoveCount,BlackMoveCount]
                EventData = SiteData = WhiteData = BlackData = ResultData = UTCDateData = UTCTimeData = WhiteEloData = BlackEloData = WhiteRatingDiffData = BlackRatingDiffData = ECOData = OpeningData = TimeControlData = TerminationData = MovesData = FirstMoveData = SecondMoveData = ThirdMoveData= WhiteMoveCount= BlackMoveCount = ''

                Eventscollected+=1
                EventInfoGathered = 0

except FileNotFoundError:
    print("The file:",local_filename, " was not found at its path.")
except Exception as e:
    print(f"An error occurred: {e}")

#Stop timer
TimeTaken = time.ctime(time.time()-TimeStart)[11:19]
print("Time Taken to load data: ",TimeTaken,"\r\n")
print("Total Events collected = "+str(Eventscollected))
print("Total lines read = "+str(LinesTotalRead))
```
